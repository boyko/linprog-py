# Network flow problems

- Node: e.g. storage facility, city, etc.
- Arc: e.g. road connection, communication line, etc.



```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()
G.add_node(1, demand=5)

G.add_node(2, demand=3)
G.add_node(3, demand=4)
G.add_edge(1, 2)

G.nodes[1]
labels = nx.get_node_attributes(G, 'demand')

plt.figure()
pos_nodes = nx.spring_layout(G)
nx.draw(G, pos_nodes, with_labels=True)

pos_attrs = {}
for node, coords in pos_nodes.items():
    pos_attrs[node] = (coords[0] + 0.12, coords[1])

node_attrs = nx.get_node_attributes(G, 'demand')
custom_node_attrs = {}
for node, attr in node_attrs.items():
    custom_node_attrs[node] = "D=" + str(attr)

plt.margins(0.15)
nx.draw_networkx_labels(G, pos_attrs, labels=custom_node_attrs)
plt.show()
```
